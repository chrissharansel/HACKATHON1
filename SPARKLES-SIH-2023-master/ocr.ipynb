```python
from PIL import Image
import pytesseract
import io
from google.colab import files
from IPython.display import Image as IPImage, display

# Set the path to the Tesseract executable (modify this according to your installation)
pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'

def extract_text_from_screenshot(image_content):
    # Convert bytes to an in-memory image
    image = Image.open(io.BytesIO(image_content))

    # Use pytesseract to extract text
    extracted_text = pytesseract.image_to_string(image)

    return extracted_text

# Allow the user to upload a PNG image
uploaded = files.upload()

# Display the uploaded image
for filename, content in uploaded.items():
    print(f'Uploaded file: {filename}')
    display(IPImage(data=content))

    # Extract text from the uploaded image
    text = extract_text_from_screenshot(content)

    print("Extracted Text:")
    print(text)
```
